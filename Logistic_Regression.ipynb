```python
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# Simple data points for logistic regression
X = np.array([[1, 2], [2, 1], [3, 3], [4, 2], [5, 4]])
y = np.array([0, 0, 1, 1, 1])  # Binary labels

# Create and fit the logistic regression model
model = LogisticRegression()
model.fit(X, y)

# Predict probabilities for the training data
y_pred_proba = model.predict_proba(X)[:, 1]

# Predict binary labels
y_pred = model.predict(X)

# Visualize the results
plt.figure(figsize=(10, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', s=100, edgecolors='k', label='Actual')
plt.scatter(X[:, 0], X[:, 1], c=y_pred_proba, cmap='coolwarm', marker='s', edgecolors='k', label='Predicted Probability')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Logistic Regression')
plt.colorbar(label='Predicted Probability')
plt.legend()
plt.show()

# Calculate accuracy
accuracy = accuracy_score(y, y_pred)
print("Accuracy:", accuracy)

# Predict for a new instance
new_instance = np.array([[3, 2]])
prediction = model.predict(new_instance)
print("Prediction for new instance:", prediction[0])
```
