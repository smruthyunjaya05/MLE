```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
# Load the breast cancer dataset
data = load_breast_cancer()
X = data.data
y = data.target
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=23)

results = []
```

```python
# SVM
clf = SVC(kernel='linear', C=3)
clf.fit(X_train, y_train)
results.append(['SVM', clf.score(X_test, y_test)])
```

```python
# KNeighborsClassifier
clf2 = KNeighborsClassifier(n_neighbors=2)
clf2.fit(X_train, y_train)
results.append(['KNeighborsClassifier', clf2.score(X_test, y_test)])
```

```python
# Logistic Regression
clf3 = LogisticRegression(max_iter=1000)
clf3.fit(X_train, y_train)
results.append(['Logistic Regression', clf3.score(X_test, y_test)])
```

```python
# Decision Tree
clf4 = DecisionTreeClassifier()
clf4.fit(X_train, y_train)
results.append(['Decision Tree', clf4.score(X_test, y_test)])
```

```python
# Random Forest
clf5 = RandomForestClassifier()
clf5.fit(X_train, y_train)
results.append(['Random Forest', clf5.score(X_test, y_test)])
```

```python
# Gradient Boosting
clf6 = GradientBoostingClassifier()
clf6.fit(X_train, y_train)
results.append(['Gradient Boosting', clf6.score(X_test, y_test)])
```

```python
# Naive Bayes
clf7 = GaussianNB()
clf7.fit(X_train, y_train)
results.append(['Naive Bayes', clf7.score(X_test, y_test)])
```

```python
# Create a DataFrame for display
df_results = pd.DataFrame(results, columns=['Algorithm', 'Accuracy'])
print(df_results)
```

```python
# Plotting the results
algorithms = df_results['Algorithm']
accuracies = df_results['Accuracy']
```

```python
plt.figure(figsize=(10, 6))
plt.bar(algorithms, accuracies, color=['blue', 'orange', 'green', 'red', 'purple', 'brown', 'pink'])
plt.xlabel('Algorithm')
plt.ylabel('Accuracy')
plt.title('Accuracy of Different Algorithms on Breast Cancer Dataset')
plt.ylim(0.8, 1.0)  # Set y-axis limits for better visualization
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for readability
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```
